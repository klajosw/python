## PlantUML renderelése képként:
Először generálj egy képet a PlantUML kódból, majd jelenítsd meg azt a Colab környezetben.




```python
from plantuml import PlantUML
from IPython.display import Image

# PlantUML szöveges kód
uml_code = """
@startuml
actor User
User -> (Login)
User -> (Search for products)
(Search for products) -> (Add to cart)
@enduml
"""

# PlantUML renderelése
plantuml = PlantUML(url='http://www.plantuml.com/plantuml/img/')

# Helyes függvényhívás a diagram generálásához
plantuml.processes(uml_code)

# UML kód fájlba írása
uml_file = '/content/diagram.uml'
with open(uml_file, 'w') as f:
    f.write(uml_code)

# PlantUML renderelése
!plantuml {uml_file}

# A generált képet a Colab-ban való megjelenítéshez
Image('/content/diagram.png')
```

## osztály diagram
* User osztály két attribútummal és két metódussal rendelkezik.
* Product osztály szintén tartalmaz attribútumokat és metódust.
* A két osztály között egy asszociációs kapcsolat látható, amely azt mutatja, hogy a User osztály több Product-ot vásárolhat (1:N kapcsolat).

```python
## osztály diagramm minta
from plantuml import PlantUML
from IPython.display import Image
# Minta osztálydiagram szöveges kód
uml_code = """
@startuml
class User {
  +String name
  +String email
  +login()
  +logout()
}

class Product {
  +String productName
  +float price
  +displayDetails()
}

User "1" *-- "0..*" Product : purchases
@enduml
"""

# PlantUML renderelése
plantuml = PlantUML(url='http://www.plantuml.com/plantuml/img/')

# Az UML diagram generálása
plantuml.processes(uml_code)

## kép létrehozása

# UML kód fájlba írása
uml_file = '/content/diagram2.uml'
with open(uml_file, 'w') as f:
    f.write(uml_code)

# PlantUML renderelése
!plantuml {uml_file}

# A generált képet a Colab-ban való megjelenítéshez

Image('/content/diagram2.png')
```

```python
## osztály diagramm minta
from plantuml import PlantUML
from IPython.display import Image
# Minta osztálydiagram szöveges kód
uml_code = """
@startuml
actor Customer
actor Admin

Customer -> (Browse Products)
Customer -> (Add Product to Cart)
Customer -> (Checkout)

Admin -> (Add New Product)
Admin -> (Remove Product)

(Browse Products) --> (View Product Details)
(Add Product to Cart) --> (Enter Shipping Information)
(Checkout) --> (Payment)

@enduml
"""


# PlantUML renderelése
plantuml = PlantUML(url='http://www.plantuml.com/plantuml/img/')

# Az UML diagram generálása
plantuml.processes(uml_code)

## kép létrehozása

# UML kód fájlba írása
uml_file = '/content/diagram3.uml'
with open(uml_file, 'w') as f:
    f.write(uml_code)

# PlantUML renderelése
!plantuml {uml_file}

# A generált képet a Colab-ban való megjelenítéshez

Image('/content/diagram3.png')
```

##Aktorok:

* Customer: A vásárló, aki az interakciókat végzi.
* Résztvevők (Participants):

* Web Browser: A böngésző, amelyet a vásárló használ.
* Web Server: A webszerver, amely kezeli a vásárló kéréseit.
* Database: Az adatbázis, amely tárolja a termékeket és a rendeléseket.
* Payment Gateway: A fizetési rendszer, amely a fizetési tranzakciókat kezeli.


##Lépések:
* A vásárló elindítja az oldalt és kér egy weboldalt a böngészőjétől.
* A böngésző elküldi a kérését a szervernek, amely adatokat kér az adatbázisból.
* A vásárló termékeket böngész, termékeket ad a kosárhoz, és végül a fizetési folyamatot indítja el.
* A fizetési tranzakció befejeződik, és az adatbázis frissíti a rendelési információkat.

```python
## osztály diagramm minta
from plantuml import PlantUML
from IPython.display import Image
# Minta osztálydiagram szöveges kód
uml_code = """
@startuml
actor Customer
participant "Web Browser" as Browser
participant "Web Server" as Server
participant "Database" as DB
participant "Payment Gateway" as Payment

Customer -> Browser : Open website
Browser -> Server : Request homepage
Server -> DB : Fetch homepage data
DB -> Server : Return homepage data
Server -> Browser : Display homepage

Customer -> Browser : Browse products
Browser -> Server : Request product list
Server -> DB : Fetch products
DB -> Server : Return products
Server -> Browser : Display products

Customer -> Browser : Add product to cart
Browser -> Server : Add to cart request
Server -> DB : Update cart
DB -> Server : Return cart updated
Server -> Browser : Display cart

Customer -> Browser : Proceed to checkout
Browser -> Server : Checkout request
Server -> Payment : Process payment
Payment -> Server : Payment success
Server -> DB : Update order status
DB -> Server : Return order confirmation
Server -> Browser : Display order confirmation

@enduml
"""

# PlantUML renderelése
plantuml = PlantUML(url='http://www.plantuml.com/plantuml/img/')

# Az UML diagram generálása
plantuml.processes(uml_code)

## kép létrehozása

# UML kód fájlba írása
uml_file = '/content/diagram4.uml'
with open(uml_file, 'w') as f:
    f.write(uml_code)

# PlantUML renderelése
!plantuml {uml_file}

# A generált képet a Colab-ban való megjelenítéshez

Image('/content/diagram4.png')
```

#Tevékenységek (Activities):

Browse Products: A vásárló termékeket böngész.
Select Product: A vásárló kiválaszt egy terméket.
Add Product to Cart: A vásárló hozzáadja a terméket a kosárhoz.
Proceed to Checkout: A vásárló a pénztárhoz lép.
Enter Shipping Information: A vásárló megadja a szállítási információkat.
Make Payment: A vásárló kifizeti a rendelést.
Confirm Order: A rendszer visszaigazolja a rendelést.
Döntési pontok (Decision points):

User logged in?: Ha a felhasználó be van jelentkezve, akkor közvetlenül folytathatja a pénztári folyamatot, egyébként először be kell jelentkeznie.

```python
## osztály diagramm minta
from plantuml import PlantUML
from IPython.display import Image
# Minta osztálydiagram szöveges kód
uml_code = """
@startuml
start

:Browse Products;
:Select Product;
:Add Product to Cart;

if (User logged in?) then (yes)
  :Proceed to Checkout;
  :Enter Shipping Information;
else (no)
  :Login;
  :Proceed to Checkout;
  :Enter Shipping Information;
endif

:Make Payment;
:Confirm Order;

stop
@enduml
"""
# PlantUML renderelése
plantuml = PlantUML(url='http://www.plantuml.com/plantuml/img/')

# Az UML diagram generálása
plantuml.processes(uml_code)

## kép létrehozása

# UML kód fájlba írása
uml_file = '/content/diagram5.uml'
with open(uml_file, 'w') as f:
    f.write(uml_code)

# PlantUML renderelése
!plantuml {uml_file}

# A generált képet a Colab-ban való megjelenítéshez

Image('/content/diagram5.png')

```
